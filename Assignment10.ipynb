05. 모든 정점 사이에 간선이 존재하는 그래프가 있다. 여러분이라면 인접 행렬 표현과 인접
리스트 표현 중 어느 것을 사용하겠는가? 그리고 그렇게 선택한 이유는 무엇인가?


=> 인접 행렬을 사용한다.
이유: 모든 정점 사이에 간선이 존재하는 완전 그래프이므로 간선 수가 최대(O(V²))이다.
이 경우 인접 리스트의 공간 이점이 없으며, 인접 행렬은 간선 존재 여부를 O(1)에 확인할
수 있어 더 효율적이다.

06. 정점들 사이에 아주 드물게 간선이 존재하는 그래프가 있다. 이런 그래프를 희소

리스트는 실제 존재하는 간선만 저장하므로 공간 효율이 높아 O(V+E)의 공간만 사용한다.

16. 다음 그래프에서 위상 정렬을 할 때, STACK을 사용한 위상 정렬 수도 코드를 쓰고,
가능한 위상 정렬 순서가 모두 몇 가지인지 쓰라.

```python
Topological Sort with STACK: 
    Create STACK 
    Create visited <- false 
    for each vertex v  
        if(!visited[v]) 
            DFS(v, visited, STACK) 
    while(!STACK.empty) 
        STACK.pop 
 
DFS(v, visited, STACK) 
    visited[v] <- true 
    for i in Adj[v] 
        if(!visited[i]) 
            DFS 
    STACK.push(v)
```

위상정렬이 되기 위해서는 b는 c보다 선행되어야 한다. 오른쪽 부분에서 가능한 경로는 d->e->f->g와 d->e->f->h로 두 가지 경우가 있다. 전체 경로에서 b가 c보다 선행되는 경우는 7C2=21가지 이므로 오른쪽 부분의 2가지 경우의 경로를 모두 고려하면 가능한 위상정렬 순서는 총 21 * 2 =42가지 이다.
